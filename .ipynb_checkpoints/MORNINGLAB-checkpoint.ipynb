```python
import pandas as pd
import numpy as np
import datetime
import warnings

warnings.filterwarnings('ignore')
warnings.filterwarnings("ignore", message="numpy.ufunc size changed")

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")

%matplotlib inline

pd.set_option('display.max_columns', None)
```

# 1

```python
data =pd.read_csv("we_fn_use_c_marketing_customer_value_analysis.csv")
```

```python
data.columns=[e.lower().replace(' ', '_') for e in data.columns]  
data['effective_to_date'] = pd.to_datetime(data.effective_to_date)
```

```python
categoricals = data.select_dtypes(np.object)
```

```python
categoricals = categoricals.set_index('customer')
```

# 2

```python
for i in categoricals.columns:
    plt.hist(categoricals[i], bins = 50)
    plt.show()
```

Some data imbalance in some columns. There might be chances to group things together.

```python
sns.histplot(data['effective_to_date'])
```

```python
data['week'] = data['effective_to_date'].apply(lambda x: x.week)
```

```python
data['week'].unique()
```

```python
data['day'] = data['effective_to_date'].apply(lambda x: x.day)
```

```python
data['day'].unique()
```

```python
data['weekday'] = data['effective_to_date'].apply(pd.Timestamp)
```

```python
data['weekday'].head()
```

```python
n, bins, patches = plt.hist(x=data['weekday'] , bins='auto', color='#0504aa',
                            alpha=0.7, rwidth=1)
plt.grid(axis='y', alpha=0.75)
plt.xlabel('effective_to_date')
plt.ylabel('count')
plt.title('My Super Very Own Histogram')
# Set a clean upper y-axis limit.
```
